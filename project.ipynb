#### Exercise 10. Are the average prices per square meter (price/m2) of "Valdemorillo" and "Galapagar" the same? (★★☆)

Print both average prices and then write a conclusion about it.

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python
import pandas as pd

# Load the dataset
data_frame = pd.read_csv('assets/real_estate.csv', sep=';')

# Ensure the 'price' and 'surface' columns are numeric to handle any invalid entries
data_frame['price'] = pd.to_numeric(data_frame['price'], errors='coerce')
data_frame['surface'] = pd.to_numeric(data_frame['surface'], errors='coerce')

# Create a new column for price per square meter (pps)
data_frame['pps'] = data_frame['price'] / data_frame['surface']

# Filter the dataset for "Valdemorillo"
valdemorillo_data = data_frame[data_frame['level5'] == 'Valdemorillo']

# Filter the dataset for "Galapagar"
galapagar_data = data_frame[data_frame['level5'] == 'Galapagar']

# Calculate the average price per square meter for Valdemorillo
valdemorillo_mean_pps = valdemorillo_data['pps'].mean()

# Calculate the average price per square meter for Galapagar
galapagar_mean_pps = galapagar_data['pps'].mean()

# Print the average prices per square meter
print(f"Average price per square meter in Valdemorillo: {valdemorillo_mean_pps:.2f}")
print(f"Average price per square meter in Galapagar: {galapagar_mean_pps:.2f}")

# Print the conclusion
if valdemorillo_mean_pps == galapagar_mean_pps:
    print("The average prices per square meter in Valdemorillo and Galapagar are the same.")
else:
    print("The average prices per square meter in Valdemorillo and Galapagar are different.")
```

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data_frame = pd.read_csv('assets/real_estate.csv', sep=';')

# Define the list of populations for the "south belt" of Madrid
south_belt_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Filter the DataFrame to include only the specified populations
south_belt_df = data_frame[data_frame['level5'].isin(south_belt_populations)]

# Normalize prices by population
south_belt_df['normalized_price'] = south_belt_df.groupby('level5')['price'].transform(
    lambda x: (x - x.min()) / (x.max() - x.min())
)

# Plot histograms for each population
plt.figure(figsize=(10, 6))

for population in south_belt_populations:
    # Filter data for the current population
    population_data = south_belt_df[south_belt_df['level5'] == population]
    
    # Plot histogram
    plt.hist(population_data['normalized_price'], bins=15, alpha=0.5, label=population)

# Add labels and title
plt.title('Normalized Price Histograms for the South Belt of Madrid')
plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
plt.legend(title='Populations')

# Show the plot
plt.tight_layout()
plt
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph. You must use the subset obtained in Exercise 14 (★★☆)
 
Hint: Make a scatter plot of each population using subplots.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data_frame = pd.read_csv('assets/real_estate.csv', sep=';')

# Define the list of populations for the "south belt" of Madrid
south_belt_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Filter the DataFrame to include only the specified populations
south_belt_df = data_frame[data_frame['level5'].isin(south_belt_populations)]

# Create subplots (2 rows and 2 columns)
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Price vs Surface Area for South Belt Populations', fontsize=16)

# List of subplots (axes) and populations to plot
population_subplot = {
    "Fuenlabrada": axes[0, 0],
    "Leganés": axes[0, 1],
    "Getafe": axes[1, 0],
    "Alcorcón": axes[1, 1]
}

# Iterate over each population and plot price vs surface
for population, ax in population_subplot.items():
    # Filter data for the current population
    population_data = south_belt_df[south_belt_df['level5'] == population]
    
    # Scatter plot of price vs surface
    ax.scatter(population_data['surface'], population_data['price'], alpha=0.6)
    
    # Set title and labels for each subplot
    ax.set_title(f"{population}", fontsize=14)
    ax.set_xlabel('Surface Area (m²)')
    ax.set_ylabel('Price (€)')

# Adjust layout for better spacing between subplots
plt.tight_layout(rect=[0, 0, 1, 0.96])

# Show the plot
plt.show()
```

```python
## Aquí: traza la coordenadas de los estados

## HERE: Plot the coordinates of the estates

import pandas as pd
import matplotlib.pyplot as plt
from ipyleaflet import Map, basemaps, Marker
from ipywidgets import Layout

# Define the map centered on Madrid (adjust the coordinates if needed)
madrid_map = Map(center = (60, -2.2), zoom = 2, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik  # OpenStreetMap as an alternative
)

# Load the dataset
data_frame = pd.read_csv('assets/real_estate.csv', sep=';')

# Define the south belt populations
south_belt_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Filter the dataset for the south belt populations
south_belt_df = data_frame[data_frame['level5'].isin(south_belt_populations)]

# Ensure correct column names (adjust based on actual dataset)
coordinates = south_belt_df[['level5', 'latitude', 'longitude']].dropna()

# Transform the coordinates into a dictionary
coordinates_dict = coordinates.to_dict('records')

# Create a scatter plot for the south belt populations
plt.figure(figsize=(10, 8))

for population in south_belt_populations:
    pop_data = coordinates[coordinates['level5'] == population]
    plt.scatter(pop_data['longitude'], pop_data['latitude'], label=population, alpha=0.7)

# Adding plot details
plt.title("Coordinates of South Belt Populations in Madrid")
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.legend(title="population")
plt.grid(alpha=0.3)
plt.show()

# Add markers to the interactive map
for location in coordinates_dict:
    marker = Marker(location=(location['latitude'], location['longitude']), draggable=False)
    madrid_map.add_layer(marker)

# Display the interactive map
madrid_map
```
